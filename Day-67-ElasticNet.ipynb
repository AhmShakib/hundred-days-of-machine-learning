```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score 

import pandas as pd 
import numpy as np 
```

```python
from sklearn.datasets import load_diabetes

data = load_diabetes() 
df = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
df.head()
```

```python
df['target'] = data.target
```

```python
df.head()
```

# Simple Linear Regression

```python
X_train,X_test,y_train,y_test = train_test_split(
    df.iloc[:,:-1],df.iloc[:,-1:],test_size = 0.2, random_state= 42)
```

```python
X_train.shape,X_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train) 

y_predlr = lr.predict(X_test)

r2_score(y_test,y_predlr)
```

# Ridge Regression 

```python
rg = Ridge(alpha = 0.1) 
rg.fit(X_train,y_train) 

y_predrg = rg.predict(X_test) 

r2_score(y_test,y_predrg)
```

# Lasso Regression 

```python
lo = Lasso(alpha = 0.1) 
lo.fit(X_train,y_train) 

y_predlo = lo.predict(X_test)

r2_score(y_test,y_predlo)
```

# ElasticNet Regression 

```python
en = ElasticNet(alpha = 0.0009,l1_ratio = 0.5) 
en.fit(X_train,y_train) 

y_preden = en.predict(X_test)  

r2_score(y_test, y_preden)
```
