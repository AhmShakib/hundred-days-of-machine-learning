```python
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report,confusion_matrix

import pandas as pd 
import numpy as np 
```

# For Binary Classification

```python
df_one = pd.read_csv('placement_class.csv')
df_one.head()
```

```python
df_one.drop((df_one.columns[0]), axis = 1, inplace= True)
```

```python
df_one.head()
```

```python
df_one.shape
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test = train_test_split(df_one.iloc[:,:2],df_one.iloc[:,-1:], test_size= 0.4, random_state= 42)
```

```python
lor = LogisticRegression() 
lor.fit(X_train,y_train)

y_predlr = lor.predict(X_test)
```

```python
dt = DecisionTreeClassifier() 
dt.fit(X_train,y_train)

y_preddt = dt.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predlr))

print(accuracy_score(y_test,y_preddt))
```

```python
print(classification_report(y_test,y_predlr))
print(classification_report(y_test,y_preddt))
```

```python
print(confusion_matrix(y_test,y_predlr))
print(confusion_matrix(y_test,y_preddt))
```

# For 3-class Dataset 

```python
import seaborn as sns 

df_two = sns.load_dataset('iris')
```

```python
df_two.head()
```

```python
from sklearn.preprocessing import LabelEncoder 

encoder = LabelEncoder() 
df_two['species'] = encoder.fit_transform(df_two['species'])
```

```python
df_two.tail(3)
```

```python
X_train1,X_test1, y_train1, y_test1 = train_test_split(df_two.iloc[:,:4],df_two.iloc[:,-1:], test_size= 0.3, random_state= 42)
```

```python
lor1 = LogisticRegression(multi_class= 'multinomial')
lor1.fit(X_train1,y_train1)
```

```python
y_predlr1 = lor1.predict(X_test1) 
```

```python
dt1 = DecisionTreeClassifier() 
dt1.fit(X_train1,y_train1) 

y_preddt1 = dt1.predict(X_test1)
```

```python
print(classification_report(y_test1,y_predlr1))

print(classification_report(y_test1,y_preddt1))
```

```python
print(confusion_matrix(y_test1,y_predlr1))         
print(confusion_matrix(y_test1, y_preddt1))      
```
